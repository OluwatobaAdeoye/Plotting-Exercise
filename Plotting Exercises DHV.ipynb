```python
# importing the packages you'll need for later
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import pyarrow

Exercises = pd.read_parquet('exercises_problem2.parquet')
```

```python
# Points
x1, y1 = 3, 5
x2, y2 = 7, 2
```

```python
# Question 1: Calculate slope (m) and y-intercept (c) of the line
m = (y2 - y1) / (x2 - x1)
c = y1 - m*x1
```

```python
# Define the line equation
def line_eq(x):
    return m*x + c
```

```python
# X range for plotting the line beyond the points
x_range = np.linspace(0, 10, 100)
```

```python
# Plot
plt.figure(figsize=(8,6))
plt.plot(x_range, line_eq(x_range), label='Line', color='black') # Line
plt.scatter([x1, x2], [y1, y2], color='black', s=100, label='Points') # Points
# Styling
plt.title('Plot of Points and the Line Through Them', fontsize=16)
plt.xlabel('X-axis', fontsize=14)
plt.ylabel('Y-axis', fontsize=14)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.legend(fontsize=14)
plt.axis([0, 10, 0, 7]) # Appropriately selected ranges

plt.show()
```

```python
#Question 2_
np.random.seed(0) # For reproducibility
dates = pd.date_range('20210101', periods=100)
category = np.random.choice(['A', 'B', 'C'], size=100)
values = np.random.randn(100).cumsum()

Exercises = pd.DataFrame({'Date': dates, 'Category': category, 'Value': values})

Exercises.head()
```

```python
# Time Series Plot
plt.figure(figsize=(10, 6))
plt.plot(Exercises['Date'], Exercises['Value'], label='Value', color='black')

# Styling
plt.title('Time Series of Values', fontsize=16)
plt.xlabel('Date', fontsize=14)
plt.ylabel('Value', fontsize=14)
plt.xticks(fontsize=12, rotation=45)
plt.yticks(fontsize=12)
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.tight_layout()  # Adjust layout to not cut off labels

plt.show()
```

```python
# Category Distribution
plt.figure(figsize=(8, 6))
Exercises['Category'].value_counts().plot(kind='bar', color='black')

# Styling
plt.title('Distribution of Categories', fontsize=16)
plt.xlabel('Category', fontsize=14)
plt.ylabel('Frequency', fontsize=14)
plt.xticks(fontsize=14, rotation=0)
plt.yticks(fontsize=14)
plt.grid(axis='y', which='both', linestyle='--', linewidth=0.5)
plt.tight_layout()

plt.show()
```

```python
# Category Trends
plt.figure(figsize=(10, 6))

# Plot each category
for category in Exercises['Category'].unique():
    subset = Exercises[Exercises['Category'] == category]
    plt.plot(subset['Date'], subset['Value'], label=f'Category {category}')

# Styling
plt.title('Value Trends by Category', fontsize=16)
plt.xlabel('Date', fontsize=14)
plt.ylabel('Value', fontsize=14)
plt.xticks(fontsize=12, rotation=45)
plt.yticks(fontsize=12)
plt.legend(fontsize=12)
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.tight_layout()

plt.show()
```
